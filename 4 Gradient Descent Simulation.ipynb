## Implementing The 1D GD Algorithm

```python
#Defining the function f(x) = x^2
def f(x):
    y = x**2
    return y

#Defining the gradient function g(x) of f(x) = derivative of f(x)
def grad(x):
    g = 2*x
    return g
```

```python
#This function calculates the gradient descent when the initial 
#value of f(x) (denoted by x0) and the learning rate (r) is passed as arguments

def calculateGD(x0, r, stoppingParameter, pri=False):
    print("\n\n\t\tWhen x0=,",x0," learning rate (r) = ",r,
          " and stopping parameter = ",stoppingParameter, ":\n", sep="")
    iteration = 0            #empty list to count the umber of iterations
    stoppingCriteriaCtr = 0  #Stopping Criteria Counter; When its value becomes 5, the algorithm will stop
    fX = []                  #empty list to store the values of f(x)
    gX = []                  #empty list to store the values of g(x) or the gradient function
    X0 = []                  #empty list to store the values of x0
    X1 = []                  #empty list to store the values of x1
    i  = []                  #empty list to store the values of iterations
    
    while stoppingCriteriaCtr <=5:
        fX.append(f(x0)) #Calculating the value of f(x)
        
        gradX = grad(x0) #Calculating and storing the value of g(x)
        gX.append(gradX) 
        
        x1 = x0-r*gradX  #Calculating the value of x1 (gradient descent (denoted by -ve sign))
        X1.append(x1)     #Storing the value of x01
        X0.append(x0)     #Storing the value of x0
        
        iteration = iteration + 1 #Calculating & storing the value of iteration number
        i.append(iteration)
        
        #In the following code, if the values of x1 and x0 are same (accuracy until the 7th decimal digit),
        #for 5 times in a row, then we can assume that we have reached the lowest value of x0 and the loop
        #must halt. If the number of iterations reach 10,000 before x1 and x0 are equal, then the loop must stop.
        if(round(x0,stoppingParameter)==round(x1,stoppingParameter)):
            stoppingCriteriaCtr = stoppingCriteriaCtr+1
        elif(iteration==10000):
            stoppingCriteriaCtr=5
        else:
            stoppingCriteriaCtr=0
        
        x0 = x1

    #Collating our findings into a data frame
    data = {"Iteration #":i, "Function f(x)":fX, "Gradient Function g(x)":gX, "Input Value (x0)":X0}
    dataDF = pd.DataFrame(data)
    
    #Print the Data Frame if the value of Print is True, else just print the last value
    if(pri==True):
        print("\n\n",dataDF)
    else:
        print("\n\nWhen Local Minima is achieved:\n\n", dataDF.iloc[-1])
    
    print("\nLocal minima is achieved in ",max(dataDF["Iteration #"]), " iterations.", sep="")
    plt.scatter(X0,fX)
    
    plt.xlabel('Value of x0', fontsize=12)
    plt.ylabel('Function f(x)', fontsize=12)
    plt.show()
```

```python
#Starting value is 2 for all the following cases
#Stopping parameter is 7
calculateGD(2,0.1,7,pri=True) #This prints all the values of f(x), g(x), Input Value(x0)

lr = [0.01,0.001,0.0001,0.00001] #Various learning rates
sp = [7,5,3]                         #Various Stopping Parameter
for i in lr:
    for j in sp:
        calculateGD(2,i,j)   #Not printing the entire table, because the output would be excruciatingly long
```

From the above code, it is clear that the number of iterations to reach the local minima:
* Increases with a decrease in learning rate, provided the stopping parameter is constant
* Decreases with a decrease in stopping parameter, provided the learning rate is constant

It is important to note here, that we calculate the difference in the x0 and x1 in each iteration and check if this difference is close to zero considering the "stopping parameter" number of decimal places. Thus, for example, if the stopping parameter is set as 7 then the value of x0 and x1, accurate upto 7 decimal places, must be equal. If this equal value appears for 5 times in a row, then this means that the value of the gradient is no longer decreasing and is constant. This is a cue for the algorithm to exit the iterations.

## Implementing The 2D and 5D GD Algorithm using Matrix

```python
#Defining the function f(x) = a + bx^2. In matrix notation it is given by f(x) = x'Ax
def MDf(x,a):
    y = x.T @ a @ x
    return y

#Defining the gradient function g(x) of f(x) = derivative of f(x). In matrix notation it is given by f(x) = 2Ax
def MDgrad(x,a):
    #g = 2 * (a @ x)
    g = (a.T + a) @ x
    return g
```

```python


#This function will work for all multi-dimension matrix such as 2D and 5D
def calculateMDGD(x0, a, r, pri=False):
    print("\n\nWhen Learning rate (r) = ",r,":\n", sep="")
    iteration = 0            #empty list to count the umber of iterations
    stoppingCriteriaCtr = 0  #Stopping Criteria Counter; When its value becomes 5, the algorithm will stop
    fX = []                  #empty list to store the values of f(x)
    gX = []                  #empty list to store the values of g(x) or the gradient function
    X0 = []                  #empty list to store the values of x0
    X1 = []                  #empty list to store the values of x1
    i  = []                  #empty list to store the values of iterations
    
    while stoppingCriteriaCtr <=10:
        fX.append(MDf(x0,a)) #Calculating the value of f(x)
        
        gradX = MDgrad(x0,a) #Calculating and storing the value of g(x)
        gX.append(gradX) 
        
        x1 = x0-r*gradX  #Calculating the value of x1 (gradient descent (denoted by -ve sign))
        X1.append(x1)     #Storing the value of x01
        X0.append(x0)     #Storing the value of x0
        
        iteration = iteration + 1 #Calculating & storing the value of iteration number
        i.append(iteration)
        
        #In the following code, if the values of x1 and x0 are same (accuracy until the 7th decimal digit),
        #for 10 times in a row, then we can assume that we have reached the lowest value of x0 and the loop
        #must halt. If the number of iterations reach 10,000 before x1 and x0 are equal, then the loop must stop.
        
        tempx0 = np.around(x0,7)
        tempx1 = np.around(x1,7)
        if(all(tempx0==tempx1)==True):
            stoppingCriteriaCtr = stoppingCriteriaCtr+1
        elif(iteration==10000):
            stoppingCriteriaCtr=10
        else:
            stoppingCriteriaCtr=0
        #stoppingCriteriaCtr = stoppingCriteriaCtr+1
        x0 = x1

    #Collating our findings into a data frame
    if(pri==True):
        data = {"Iteration #":i, "Function f(x)":fX, "Gradient Function g(x)":gX, "Input Value (x0)":X0}
        dataDF = pd.DataFrame(data)
        print("\n\n",dataDF)
    
    print("Point of convergence: \n",x1, sep="")
    print("Local minima is achieved in ",iteration, " iterations.", sep="")

    X=np.asarray(X1)[:,0]
    Y=np.asarray(X1)[:,1]
    return(X,Y)
    
```

```python
#This function plots the contour plot of the gradient descent of the function defined above
#This function also plots the various values of x0 (both X and Y values) that it takes over all the iterations
#This function takes in a "window parameter" that adjusts the size of the plot as required
#This function also takes in the matrix a as a parameter that is later passed to f(x,a)
def plotContour(X,Y,a,windowParameter):
    ex1 = np.linspace(-windowParameter, windowParameter, 30) #Equally spaced 30 numbers
    ex2 = np.linspace(-windowParameter, windowParameter, 30) #Equally spaced 30 numbers
    (xm1, xm2) = np.meshgrid(ex1, ex2) 
    # check what is the shape of xm1, xm2!
    #x1 = xm1.ravel() ravel function just flattens the array. This is useless here.
    # check, what exactly does 'ravel' do
    #x2 = xm2.ravel()
    z = np.empty_like(xm1)
    ## now just do a nested loop
    for i in range(30):
        for j in range(30):
            x = np.array([[xm1[i,j]], [xm2[i,j]]]) 
            #Here the correct format is 2x1 array. MDf(x,a), the dimension of x must be 2x1.
            z[i,j] = MDf(x,a)
    ## and contour plot it:
    fig = plt.figure(figsize=(5,5))
    ax = fig.add_subplot(111)
    ax.set_aspect('equal')
    # use equal aspect ratio: the gradient will show up correctly
    cs = ax.contour(ex1, ex2, z)
    # create the contour plot
    ax.clabel(cs, inline=1, fontsize=10)
    # you may want to label the contour, for your reference and debugging.
    plt.scatter(X,Y, c='red', s=5)
    plt.plot(X,Y, c='red')
    plt.show()
```

```python
x = np.array([[2],[-3]])
a = np.array([[1,0],[0,1]])

print("\nWhen X0 = \n", x,"\n")

X,Y = calculateMDGD(x, a, 0.1)
plotContour(X,Y,a,3)

X,Y = calculateMDGD(x, a, 0.01)
plotContour(X,Y,a,3)

X,Y = calculateMDGD(x, a, 0.001)
plotContour(X,Y,a,3)
```

From the above analyses we can ascertain that:
* As the learning rate (r) decreases, the number of iterations to achieve the local minima/convergence increases.

```python
A = np.arange(1,26).reshape(5,5)
A = 0.5 * (A.T + A) + 10*np.identity(5)
x0 = np.array([[10],[20],[30],[40],[50]])
print("\nWhen X0 = \n", x0,"\n")

#X,Y = calculate5DGD(x0, A, 0.1) #This gives a runtime error #
lr = [0.01, 0.001, 0.0001]
for i in lr:
    X,Y = calculateMDGD(x0, A, i)
```

## Condition Numbers

```python
A = np.array([[1,1],[1,1]])
print("\nDeterminant:",np.linalg.det(A)) #Calculating the determinant
print("\nCondition Number:",np.linalg.cond(A))
```

* Since the determinant of the above matrix is 0, hence it is singular
* SInce the matrix is singular, its condition number is infinite.

```python
def conditionNumbers(alpha,A):
    c = A + alpha*np.eye(2)
    ev = np.linalg.eig(c) #eigen value
    cn = np.linalg.cond(c) #condition numbers
    print("\n\n\t\tWhen alpha =",alpha)
    print("\nThe unit matrix becomes:\n",c)
    print("\nEigen Values:\n",ev)
    print("\nCondition Number:",cn)
    print("\nDeterminant:",np.linalg.det(c))
    
```

```python
cn = [0.01,1,100]
for i in cn:
    conditionNumbers(i,A)
```

* Condition number of a matrix A is defined as ||A|| * ||A^-1|| where ||A|| is the norm of matrix A
* If a matrix is singular, then its condition number is infinite.
* The condition number of the matrix measures the ratio of the maximum relative stretching to the maximum relative shrinking that matrix does to any non zero vectors.
* Another way to say that the condition number of a matrix measures the amount of distortion of the unit sphere (in the corresponding vector norm) under the transformation by the matrix. The larger the condition number, the more distorted (relatively long and thin) the unit sphere becomes when transformed by the matrix.
* Source: http://www.cse.iitd.ernet.in/~dheerajb/CS210_lect07.pdf

```python
def difAlpha(alpha,A):
    c = A + alpha*np.eye(2)
    return c
```

```python
A = np.array([[1,1],[1,1]])
x0 = np.array([2,-3])

alpha = [0.01,1,100]
lr = [0.1, 0.01, 0.001, 0.0001]

for i in alpha:
    a = difAlpha(i,A)
    for j in lr:
        if(i==100 and (j==0.1 or j==0.01)): 
            continue
            #We are excluding the case where alpha = 100 and learning rate is either 0.1 or 0.01 
            #because in these cases we are unable to reach the minima as we overshoot it
        else:
            X,Y = calculateMDGD(x0, a, j)
            conditionNumbers(i,a)
            plotContour(X,Y,a,3)
```

* For constant alpha, the number of iterations/time taken for convergence increase with a decrease in learning rate
* For constant alpha, the condition number remains constant when the learning rate decreses
* For constant learning rate, the number of iterations/time taken for convergence decrease with a increase in alpha
* For constant learning rate, the condition number decreases drastically with a increase in alpha

## Loss function of linear regression

```python
x = np.random.uniform(size=(100,1)) 
X = np.concatenate((np.ones_like(x), x), axis=1)
eps = np.random.normal(size=(100,1))
betaTrue = np.array([[1],[1]])
yTrue = (X @ betaTrue) + eps

values = (np.linalg.inv((X.T @ X))) @ X.T @ yTrue
print("\n\nTrue Eigen Values or true Betas of the equation:\n",values)
```

```python
#Calculating the loss function
def lossFunction(beta):
    yPredicted = X @ beta
    error = yPredicted - yTrue
    lossFun = error.T @ error
    return(lossFun)

def gradLossFunction(beta):
    yPredicted = X @ beta
    gradLossFun = X.T @ (2*(yPredicted - yTrue)) 
    return(gradLossFun)


def calculateLRGD(x0, r, pri=False):
    print("\n\nWhen x0=\n",x0," and learning rate (r) = ",r,":\n", sep="")
    iteration = 0            #empty list to count the umber of iterations
    stoppingCriteriaCtr = 0  #Stopping Criteria Counter; When its value becomes 5, the algorithm will stop
    fX = []                  #empty list to store the values of f(x)
    gX = []                  #empty list to store the values of g(x) or the gradient function
    X0 = []                  #empty list to store the values of x0
    X1 = []                  #empty list to store the values of x1
    i  = []                  #empty list to store the values of iterations
    
    while stoppingCriteriaCtr <=10:
        fX.append(lossFunction(x0)) #Calculating the value of f(x)
        
        gradX = gradLossFunction(x0) #Calculating and storing the value of g(x)
        gX.append(gradX) 
        
        #print("\n\n",gradX)
        #print("\n",x0)
        
        x1 = x0-r*gradX  #Calculating the value of x1 (gradient descent (denoted by -ve sign))
        X1.append(x1)     #Storing the value of x01
        X0.append(x0)     #Storing the value of x0
        
        iteration = iteration + 1 #Calculating & storing the value of iteration number
        i.append(iteration)
        
        #In the following code, if the values of x1 and x0 are same (accuracy until the 7th decimal digit),
        #for 10 times in a row, then we can assume that we have reached the lowest value of x0 and the loop
        #must halt. If the number of iterations reach 10,000 before x1 and x0 are equal, then the loop must stop.
        
        tempx0 = np.around(x0,7)
        tempx1 = np.around(x1,7)
        if((tempx0==tempx1).all()==True):
            stoppingCriteriaCtr = stoppingCriteriaCtr+1
        elif(iteration==100000):
            stoppingCriteriaCtr=10
        else:
            stoppingCriteriaCtr=0
       
        #stoppingCriteriaCtr = stoppingCriteriaCtr+1
        x0 = x1

    #Collating our findings into a data frame
    if(pri==True):
        data = {"Iteration #":i, "Function f(x)":fX, "Gradient Function g(x)":gX, "Input Value (x0)":X0}
        dataDF = pd.DataFrame(data)
        print("\n\n",dataDF)
    
    print("Point of convergence:\n",x1, sep="")
    print("Local minima is achieved in ",iteration, " iterations.", sep="")

    X=np.asarray(X1)[:,0]
    Y=np.asarray(X1)[:,1]
    return(X,Y)
```

```python
#This function plots the contour plot of the gradient descent of the function defined above
#This function also plots the various values of x0 (both X and Y values) that it takes over all the iterations
#This function takes in a "window parameter" that adjusts the size of the plot as required
#This function also takes in the matrix beta as a parameter that is later passed to lossFunction(beta)
def plotContour(X,Y, windowParameter):
    ex1 = np.linspace(-windowParameter, windowParameter, 30)
    ex2 = np.linspace(-windowParameter, windowParameter, 30)
    (xm1, xm2) = np.meshgrid(ex1, ex2)
    # check what is the shape of xm1, xm2!
    #x1 = xm1.ravel()
    # check, what exactly does 'ravel' do
    #x2 = xm2.ravel()
    z = np.empty_like(xm1)
    ## now just do a nested loop
    for i in range(30):
        for j in range(30):
            x = np.array([[xm1[i,j]], [xm2[i,j]]])
            z[i,j] = lossFunction(x)
    ## and contour plot it:
    fig = plt.figure(figsize=(5,5))
    ax = fig.add_subplot(111)
    ax.set_aspect('equal')
    # use equal aspect ratio: the gradient will show up correctly
    cs = ax.contour(ex1, ex2, z)
    # create the contour plot
    ax.clabel(cs, inline=1, fontsize=10)
    # you may want to label the contour, for your reference and debugging.
    plt.scatter(X,Y, c='red', s=5)
    plt.plot(X,Y, c='red')
    plt.show()
```

```python
beta = np.array([[20],[-18]])
lr = [0.0001,0.00001,0.000001]
for i in lr:
    P,Q = calculateLRGD(beta, i)#, True
    plotContour(P,Q,20)
```

By implementing the Gradient Descent for linear regression above, we can observe that:
* For larger learning rates (0.1, 0.001 and 0.0001) the algorithm stops working. This is because at that learning rate, the original value of beta is alreay low and it cannot be lower than that in subsequent iterations
* As learning rate decreases, the point of convergence is reached slower as the number of iterations increases
* We can see that the above algorithm is implemented correctly because the point of convergence perfectly coincides with the eigen values calculated separately.

* Thus, I guess it is a better approach to increase the initial/starting learning rate by a certain percentage in each iteration to make the gradient converge faster.
